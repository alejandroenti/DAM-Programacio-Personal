## Exercici 0

Fes una funció (calcula_interessos) que calculi els interessos que es paga per una hipotèca, segons un capital inicial, un interès i un nombre de mesos.

Per exemple, si el capital inicial és de 1000€, l'interès del 5% i el nombre de mesos 10, el resultat hauria de ser 50€.

**Pista**: Els interessos es calculen com a `capital * interès * mesos / 100`.

Comprova que la funció 'calcula_interessos' funciona correctament amb els següents valors: 

* 1000, 5, 10 => 500
* 2500, 3, 12 => 900
* 7464, 4, 14 => 4179.84
* 10000, 2, 24 => 4800

````python
# Resol aquí l'exercici 0
def calcula_interessos(init_capital, interests, num_months):
    return (init_capital * interests * num_months) / 100

print(f"Calcula interessos amb 1000, 5, 10: {calcula_interessos(1000, 5, 10)}")
print(f"Calcula interessos amb 2500, 3, 12: {calcula_interessos(2500, 3, 12)}")
print(f"Calcula interessos amb 7464, 4, 14: {calcula_interessos(7464, 4, 14)}")
print(f"Calcula interessos amb 10000, 2, 24: {calcula_interessos(10000, 2, 24)}")
````

## Exercici 3

Fes una funció (valida_contrasenya) que a partir d'una contrasenya comprovi que compleix els següents requisits:

* Tenir entre 5 i 10 caràcters (ambdós inclosos)
* Tenir almenys dues lletres majúscules
* Tenir almenys tres lletres minúscules
* Tenir almenys dos números

La funció ha de retornar `True` si la contrasenya és vàlida i `False` si no ho és.


Aleshores fes un programa que demani una contrasenya i comprovi si és vàlida o no.

* Si la contrasenya és vàlida el programa mostra: "Contrasenya vàlida"
* Si no ho és mostra: "Contrasenya no vàlida".

````python
# Fes aquí el codi de l'exercici 3
def valida_contrasenya(password):
    length = len(password)
    num_majus = len(list(filter(str.isupper, password)))
    num_minus = len(list(filter(str.islower, password)))
    num_nums = len(list(filter(str.isnumeric, password)))

    return (5 <= length <= 10) and num_majus >= 2 and num_minus >= 3 and num_nums >= 2

password = input("Introdueix la teva contrasenya: ")

if valida_contrasenya(password):
    print("Contrasenya vàlida")
else:
    print("Contrasenya no vàlida")
````

Escriu una funció (calcula_consum) que determini la factura d'aigua d'un client segons el nombre de litres que ha consumit durant un mes.

```text
Cost fix mensual: 10€ (inclou els primers 1.000 litres)
Fins a 5.000 litres (inclòs): 0,02€ per litre addicional.
Fins a 10.000 litres (inclòs): 0,03€ per litre addicional.
Més de 10.000 litres: 0,05€ per litre addicional.
```
Per exemple:
* Si un client ha consumit 800 litres, hauria de pagar només el cost fix de 10€.
* Si un client ha consumit 6.000 litres: 
```text
    10€ : Cost fix (1.000 litres)
+   80€ : 0,02€ per 4.000 litres
    30€ : 0,03€ per 1.000 litres
-------            ------
   120€ : Total     6.000 litres
```
* Si et cal, fes altres funcions de suport per a calcula_consum. Sigues organitzat/da!

Exemples per comprovar:
```text
* 800 litres => 10€
* 6.000 litres => 120€
* 9.000 litres => 210€
* 12.000 litres => 340€
```

````python
# Resol aquí l'exercici 6
def calcula_consum(total_consum):
    total_to_pay = 10

    if total_consum < 1000:
        return total_to_pay
    
    if total_consum <= 5000:
        to_calculate = (total_consum - 1000) * 0.02
        total_to_pay = total_to_pay + to_calculate
        return total_to_pay
    if total_consum <= 10000:
        to_calculate = ((total_consum - 5000) * 0.03) + (4000 * 0.02)
        total_to_pay = total_to_pay + to_calculate
        return total_to_pay
    else:
        to_calculate = ((total_consum - 10000) * 0.05) + (5000 * 0.03) + (4000 * 0.02)
        total_to_pay = total_to_pay + to_calculate
        return total_to_pay

print(f"El total a pagar per 800 litres és {calcula_consum(800):.2f}€")
print(f"El total a pagar per 6000 litres és {calcula_consum(6000):.2f}€")
print(f"El total a pagar per 9000 litres és {calcula_consum(9000):.2f}€")
print(f"El total a pagar per 12000 litres és {calcula_consum(12000):.2f}€")
````

```text

Dijous (Dia familiar): Si un adult compra una entrada, l'entrada del primer nen té un 50% de descompte els altres nens no paguen.
```

## Exercici 8 

Un petit hotel a la costa vol informatitzar el seu sistema de reserves. 

Necessiten una funció (calcula_reserva) que determini el preu total d'una reserva segons el nombre de nits, el tipus d'habitació i si el client vol esmorzar inclòs.

Les tarifes són:
```text
Habitació estàndard: 50€ per nit.
Habitació amb vistes al mar: 70€ per nit.
Suite amb jacuzzi: 120€ per nit.
Esmorzar: 10€ per dia i persona.
```
La funció hauria d'acceptar els següents paràmetres: 
```text
* nombre de nits
* tipus d'habitació (estàndard, amb vistes o suite)
* el nombre de persones que volen esmorzar.
```
A més, l'hotel ofereix les següents promocions:
* Si reserves més de 5 nits, t'ofereixen un descompte del 10% sobre el total de la reserva (sense comptar l'esmorzar).
* Si reserves una suite amb jacuzzi per més de 3 nits, l'esmorzar és gratuït.

Et pot anar bé definir primer les següents funcions:
* calcula_cost_habitacio(tipus_habitacio, nits)
* calcula_cost_esmorzar(nombre_persones, nombre_dies)

Per exemple:
* Reserva d'una habitació estàndard per 4 nits sense esmorzar: 50€ * 4 = 200€.
* Reserva d'una habitació amb vistes al mar per 6 nits amb esmorzar per a 2 persones: [(70€ * 6) - 10%] + (10€ * 6 * 2) = 444€.
* Si et cal, fes altres funcions de suport per a calcula_reserva. Sigues organitzat/da!

Exemples per comprovar:
* Habitació estàndard, 4 nits, 2 persones amb esmorzar: 280€
* Suite amb jacuzzi, 5 nits, 2 persones amb esmorzar (gratuït per la promoció): 640€
* Habitació amb vistes, 6 nits, 1 persona sense esmorzar: 438€

````python
# Resol aquí l'exercici 8
def calcula_cost_habitació(tipus_habitació, nits):
    if tipus_habitació == 1:
        return 50 * nits
    if tipus_habitació == 2:
        return 70 * nits
    if tipus_habitació == 3:
        return 120 * nits

def calcula_cost_esmorzar(num_persones, num_dies):
    return 10 * num_persones * num_dies

def calcula_reserva(num_nits, tipus_habitació, num_persones):
    total_to_pay = 0
    
    if num_nits > 5:
        total_to_pay += calcula_cost_habitació(tipus_habitació, num_nits) * 0.9
        total_to_pay += calcula_cost_esmorzar(num_persones, num_nits)
        return total_to_pay
    
    if tipus_habitació == 3 and num_nits > 3:
        total_to_pay += calcula_cost_habitació(tipus_habitació, num_nits)
        return total_to_pay
    
    total_to_pay += calcula_cost_habitació(tipus_habitació, num_nits)
    total_to_pay += calcula_cost_esmorzar(num_persones, num_nits)
    return total_to_pay

print(f"Habitació estàndard, 4 nits, 2 persones per esmorzar: {calcula_reserva(4, 1, 2):.2f}€")
print(f"Suite amb jaccuzi, 5 nits, 2 persones per esmorzar: {calcula_reserva(5, 3, 2):.2f}€")
print(f"Habitació amb vistes al mar, 6 nits, 1 persones per esmorzar: {calcula_reserva(6, 2, 1):.2f}€")
````

````python
# Fes aquí el codi de l'exercici 9
def calcula_cost_lloc(tipus_lloc):
    if tipus_lloc.lower() == "sala estàndard":
        return 100
    if tipus_lloc.lower() == "jardí amb piscina":
        return 200
    if tipus_lloc.lower() == "saló gran amb escenari":
        return 500

def calcula_cost_mejar(tipus_menjar, num_convidats):
    if tipus_menjar.lower() == "menú bàsic":
        return 15 * num_convidats
    if tipus_menjar.lower() == "menú premium":
        return 30 * num_convidats

def calcula_cost_entreteniment(tipus_entreteniment):
    if tipus_entreteniment.lower() == "màgia":
        return 250
    if tipus_entreteniment.lower() == "música en directe":
        return 500
    
def calcula_festa(tipus_lloc, num_convidats, tipus_menjar, tipus_entreteniment):
    total_to_pay = 0

    if tipus_lloc.lower() == "saló gran amb escenari" and tipus_entreteniment.lower() == "música en directe":
        total_to_pay += calcula_cost_lloc(tipus_lloc)
        total_to_pay += calcula_cost_entreteniment(tipus_entreteniment) - 100
        if num_convidats > 50:
            total_to_pay += calcula_cost_mejar(tipus_menjar, num_convidats) * 0.95
        else:
            total_to_pay += calcula_cost_mejar(tipus_menjar, num_convidats)
        return total_to_pay
    
    total_to_pay += calcula_cost_lloc(tipus_lloc)
    total_to_pay += calcula_cost_entreteniment(tipus_entreteniment)
    if num_convidats > 50:
        total_to_pay += calcula_cost_mejar(tipus_menjar, num_convidats) * 0.95
    else:
        total_to_pay += calcula_cost_mejar(tipus_menjar, num_convidats)
    return total_to_pay

print(f"Jardí amb piscina, 40 convidats, menú bàsic, música en directe: {calcula_festa("jardí amb piscina", 40, "menú bàsic", "música en directe"):.2f}€")
print(f"Saló gran amb escenari, 70 convidats, menú premium, música en directe: {calcula_festa("saló gran amb escenari", 70, "menú premium", "música en directe"):.2f}€")
print(f"Sala estàndard, 15 convidats, menú premium, màgia: {calcula_festa("sala estàndard", 15, "menú premium", "màgia"):.2f}€")
````
